Diccionarios

Las claves pueden ser **de cualquier tipo inmutable**

La clave **es un valor que se usa para buscar una entrada en un diccionario**

los valores pueden ser **de cualquier tipo**

Un diccionario **establece una correspondencia entre claves y valores.**

```python
Tupla_test = tuple('Python')
print(Tupla_test)
print("La longitud es: {}".format(len(Tupla_test)))
```

```python
Tupla_ones = (1,)
tamanio=10
Tupla_ones = tamanio * Tupla_ones
print(Tupla_ones)
```

```python
tupla_hijo = ('1','2','3')
tupla_hija = ('a','b','c')
tupla_madre = (tupla_hijo+tupla_hija)
print(len(tupla_madre))
```

```python
dir_5={'Provincia':'Misiones','Ciudad:':'Obera','Calle':'Sargento Cabral','Numero':618}

if(dir_5['Provincia']=='Misiones'):
  print("Encontrado")

if(dir_5.values()=='Misiones'):
  print("Encontrado")

if(dir_5.items()=='Misiones'):
  print("Encontrado")
```
