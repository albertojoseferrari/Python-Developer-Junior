1. Crear una tupla con 10 valores.

2. Mostrar la suma de sus elementos y el promedio.

3. Además se pide averiguar si el cero esta  o no dentro de la estructura

```python
lista=[]
for i in range(10):
  elemento=int(input(f"Ingrese el {i+1}° valor: "))
  lista.append(elemento)
tupla=tuple(lista)
suma=0
bandera=0
for i in tupla:
  suma+=i
  if i==0:
    bandera=1
print("La suma de sus elementos es", suma)
print("El promedio de sus elementos es", suma/len(tupla))
if bandera==1:
  print("El cero está dentro de la estructura.")
else:
  print("El cero no está dentro de la estructura.")
```

Crear un diccionario con N valores donde la clave sea el nombre del usuario y el valor sea el teléfono.

N es una constante determinada por el usuario. No se podrán cargar nombres repetidos.

Mostrar el diccionario

```python
N=int(input("Ingrese N: "))
diccio={}
for i in range(N):
  bandera=0
  while bandera==0:
    bandera=1
    clave=input("Ingrese el nombre: ")
    for i in diccio:
      if clave==i:
        bandera=0
        print("No se pueden cargar nombres repetidos.")
  valor=int(input("Ingrese el teléfono: "))
  diccio[clave]=valor
print(diccio)
```

Solución de Grootlam:

```python
agenda = {}
N = int(input("Ingrese la cantidad de contactos que desea agregar a la agenda: "))

for i in range(N):
    nombre = input("Ingrese el nombre del contacto: ")

    for j in agenda:
        while j == nombre:
            nombre = input("El contacto ya se ha cargado. Vuelva a ingresar el nombre del contacto: ")

    telefono = input(f"Ingrese el telefono de {nombre}: ")

    while telefono.isnumeric() == False:
        telefono = input("El telefono debe poseer SÓLO numeros, intente nuevamente: ")
        if telefono.isnumeric() == True:
            break

    agenda[nombre] = telefono

print(agenda)
```

Dado el diccionario:

datos = {'nombre': 'Sergio', 'edad': 25,'Genero':'M', 'estado': 'casado'}

1.      Eliminar el segundo elemento

2.      Ingresar por teclado dos pares mas clave:valor

2.      Mostrar por pantalla sólo los valores del diccionario

3.      Recorrer el diccionario y mostrar los elementos clave:valor

```python
datos = {'nombre': 'Sergio', 'edad': 25,'Genero':'M', 'estado': 'casado'}
datos.pop("edad")
print(datos)
for i in range(2):
  clave=input("Ingrese una categoría: ")
  valor=input("Ingrese un valor: ")
  datos[clave]=valor
print("Los valores del diccionario son: ")
for i in datos.values():
  print(i)
print("Los elementos clave:valor del diccionario son: ")
for i in datos.items():
  print(i)
```

Dada la tupla TL= (1, 4, 5, 6, 3, 25,1,1,4,3)

1.      Ingresar un valor de índice por teclado y eliminar el elemento de esa posición

2.      Mostrar la tupla modificada

3.      Mostrar el valor máximo de la tupla

```python
TL= (1, 4, 5, 6, 3, 25,1,1,4,3)
indice=int(input("Ingrese un índice: "))
lista=list(TL)
del lista[indice]
TL=tuple(lista)
print("La tupla modificada es", TL)
print("El valor máximo de la tupla es", max(lista))
```
