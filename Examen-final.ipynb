# CÓDIGO

Un colegio desea almacenar las notas de N alumnos.

Se pide realizar un programa para control de los mismos. Para ello se presentará el siguiente menú:

1.       Carga de datos

2.       Búsqueda

3.       Calculo

4.       Salir

OBSERVACION: Tener en cuenta que N es un valor que se ingresa por teclado en el programa principal. Validar que se ingrese un valor positivo, sino reingresar.

Por ejemplo: si N toma el valor 5 significa que se ingresarán 5 datos de alumnos.

1.       Si elige la opción 1: se realiza con una función de nombre CARGA la cual recibirá como parámetro el valor de la cantidad de N.

La carga se hace una sola vez, es decir que se deberán cargar TODOS los datos de los N alumnos.

Los datos de cada uno de los N alumnos se cargarán en un diccionario son:

Legajo del alumno, Apellido y Nombre, valor de Nota1 y valor de Nota2.
Validar que el legajo del alumno y los valores de las notas que se ingresen sea positivos, de lo contrario, reingresar.



Ayuda: utilizar while para el reingreso de datos erróneos.

El diccionario  debe tener como clave el legajo del alumno  y como datos una lista por cada alumno. Por ejemplo:

Alumno={224:[Gonzalez Mario, 10,10], 321:[Gomez Maria, 5,7], 45:[Perez Jose, 5,9]}

Si elige la opción 2: Esta opción deberá llamar a una función de nombre BUSCA la cual recibirá como parámetro un valor de legajo del alumno y  el diccionario. El legajo del alumno a buscar se ingresa en el programa principal y es pasado como parámetro a la función.

La función debe recorrer y  buscar en el diccionario el legajo. Si el legajo está guardado en el diccionario deberá mostrar los datos correspondientes a ese legajo. Si no lo encuentra mostrar el mensaje: “El legajo XXX no se encuentra registrado”.

OBSERVACION: solo se puede elegir la opción 2 si previamente se eligió el punto 1, de lo contrario no se tendrían datos para buscar.

Sugerencia: puede utilizar una variable bandera (o lo que considere necesario)  para poder saber si se hizo o no la carga.

2.       Si elige la opción 3: esta opción llamara a una función de  nombre MUESTRA que mostrará los registros almacenados en el diccionario de a un registro por renglón.

OBSERVACION: solo se puede elegir la opción 3 si previamente se eligió el punto 1, de lo contrario no se tendrían datos para buscar

3.       Si elige la opción 4: Solo se sale del programa cuando elige 4.

ATENCION: SE PUEDEN ELEGIR LA OPCIONES DEL MENU LAS VECES QUE SE DESEE MIENTRAS NO SE PRESIONE LA OPCION 4 DE SALIDA.

```python
# FUNCIONES A UTILIZAR:

def Carga(cantidad_de_alumnos):

  for i in range(cantidad_de_alumnos):
    legajo = int(input(f"Ingrese el Legajo del {i+1}° alumno/a: "))

    # Validación del legajo:
    while legajo <= 0:
      print("El Legajo del alumno/a debe ser positivo.")
      legajo = int(input(f"Ingrese nuevamente el Legajo del {i+1}° alumno/a: "))

    apellido_y_nombre = input(f"Ingrese Apellido y Nombre del {i+1}° alumno/a: ")

    nota1 = float(input(f"Ingrese la Nota 1 del {i+1}° alumno/a: "))

    # Validación de la nota1:
    while nota1 <= 0:
      print("La Nota 1 del alumno/a debe ser positiva.")
      nota1 = float(input(f"Ingrese nuevamente la Nota 1 del {i+1}° alumno/a: "))

    nota2 = float(input(f"Ingrese la Nota 2 del {i+1}° alumno/a: "))

    # Validación de la nota2:
    while nota2 <= 0:
      print("La Nota 2 del alumno/a debe ser positiva.")
      nota2 = float(input(f"Ingrese nuevamente la Nota 2 del {i+1}° alumno/a: "))

    Alumnos[legajo]=[apellido_y_nombre,nota1,nota2]

def Busca(legajo_a_buscar,Alumnos):

  bandera_busca = 0

  for legajo in Alumnos:
    if legajo == legajo_a_buscar:
      print(f"El alumno/a de Legajo {legajo} es {Alumnos[legajo][0]} y sus notas son {Alumnos[legajo][1]} y {Alumnos[legajo][2]}.")

      # Registro que se encontró el legajo:
      bandera_busca=1

  # Si el legajo no se encontró:
  if bandera_busca == 0:
    print(f"El Legajo {legajo_a_buscar} no se encuentra registrado.")

def Muestra():

  for legajo in Alumnos:
    print(f"El alumno/a de Legajo {legajo} es {Alumnos[legajo][0]} y sus notas son {Alumnos[legajo][1]} y {Alumnos[legajo][2]}.")

# PROGRAMA PRINCIPAL:

Alumnos={}

N = int(input("Ingrese la cantidad de alumnos/as a almacenar: "))

# Validación de N:
while N <= 0:
  print("La cantidad de alumnos/as debe ser positiva.")
  N = int(input("Ingrese nuevamente la cantidad de alumnos/as a almacenar: "))

bandera_opcion1 = 0

while True:

  print('''
        MENÚ:

        1. Carga de datos
        2. Búsqueda
        3. Muestra
        4. Salir

        ''')

  opcion = input("Elija una opción: ")

  # Opción 1 (Carga de datos):
  if opcion == "1":

    Carga(N)

    # Registro que se accedió a la opción 1:
    bandera_opcion1=1

  # Opción 2 (Búsqueda):
  elif opcion == "2":

    # Chequeo que se haya elegido la opción 1 previamente:
    if bandera_opcion1 == 0:
      print("Previamente debe realizar la carga de datos (opción 1).")

    else:
      legajo = int(input(f"Ingrese el Legajo del alumno/a a buscar: "))
      Busca(legajo,Alumnos)

  # Opción 3 (Muestra):
  elif opcion == "3":

    # Chequeo que se haya elegido la opción 1 previamente:
    if bandera_opcion1 == 0:
      print("Previamente debe realizar la carga de datos (opción 1).")

    else:
      Muestra()

  # Opción 4 (Salir):
  elif opcion == "4":

    print("Usted eligió Salir.")
    break

  # Otra opción:
  else:

    print("La opción ingresada no es correcta.")
```
