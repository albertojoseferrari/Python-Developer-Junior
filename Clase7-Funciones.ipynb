1.   Escriba un programa que por medio de una función pida la anchura y altura de un rectángulo y lo dibuje con caracteres producto (*).

```python
def rectangulo(ancho,alto):
  print(ancho*"*")
  for i in range(2,alto):
    print("*",(ancho-4)*" ","*")
  print(ancho*"*")

rectangulo(7,4)
```

```python
def rectangulo(ancho,alto,caracter):
  print(ancho*caracter)
  for i in range(2,alto):
    print(caracter,(ancho-4)*" ",caracter)
  print(ancho*caracter)

rectangulo(7,4,"°")
```

3. Escriba un programa que permita que una función reciba una lista de valores (que puede ser vacía).
El programa tiene que pedir un número y luego solicitar ese número de valores para crear la lista. Luego, a traves de otra función devolver al programa principal la lista y el mayor y menor valor de la misma.

```python
def armado(N):
  valores=[]
  for i in range(N):
    elemento=int(input(f"Ingrese el {i+1}° valor de la lista: "))
    valores.append(elemento)
  return valores

def extremos(lista_armada):
  maximo=max(lista_armada)
  minimo=min(lista_armada)
  return maximo,minimo

cantidad=int(input("Ingrese el número de valores de la lista: "))
lista=armado(cantidad)
a,b=extremos(lista)
print(f"La lista es {lista}, el mayor valor es {a} y el menor valor es {b}.")
```

4. Crear una función que calcule la temperatura media de un día a partir de la temperatura máxima y mínima. Crear un programa principal, que utilizando la función anterior, vaya pidiendo la temperatura máxima y mínima de cada día y vaya mostrando la media. El programa pedirá el número de días que se van a introducir.

```python
def temp_media(maxima,minima):
  return (maxima+minima)/2

N=int(input("Ingrese el número de días que se van a introducir: "))
for i in range(N):
  maxima_del_dia_i = int(input(f"Ingrese la temperatura máxima del día {i+1}: "))
  minima_del_dia_i = int(input(f"Ingrese la temperatura mínima del día {i+1}: "))
  temp_media_del_dia_i = temp_media(maxima_del_dia_i,minima_del_dia_i)
  print(f"La temperatura media del día {i+1} es {temp_media_del_dia_i}.")
```

5. Crear una función llamada “Login”, que recibe un nombre de usuario y una contraseña y te devuelve Verdadero si el nombre de usuario es “usuario1” y la contraseña es “asdasd”. Además recibe el número de intentos que se ha intentado hacer login y si no se ha podido hacer login incremente este valor.

```python
def Login(usuario,contrasenia,intentos):
  if usuario != "usuario1" or contrasenia != "asdasd":
    intentos += 1
  return usuario == "usuario1" and contrasenia == "asdasd", intentos

Login("usuario2","asdasd",3)
```

6. Crear un programa principal donde se pida un nombre de usuario y una contraseña y se intente hacer login, solamente tenemos tres oportunidades para intentarlo.

```python
def Login(usuario,contrasenia,intentos):
  if usuario != "usuario1" or contrasenia != "asdasd":
    intentos += 1
  return usuario == "usuario1" and contrasenia == "asdasd", intentos

usuario_ingresado = ""
contrasenia_ingresada = ""
intentos_realizados=0
while not Login(usuario_ingresado,contrasenia_ingresada,intentos_realizados)[0] and Login(usuario_ingresado,contrasenia_ingresada,intentos_realizados)[1]<=3:
  usuario_ingresado=input("Usuario: ")
  contrasenia_ingresada=input("Contraseña: ")
  if Login(usuario_ingresado,contrasenia_ingresada,intentos_realizados)[0]:
    print("Login correcto.")
  else:
    intentos_realizados=Login(usuario_ingresado,contrasenia_ingresada,intentos_realizados)[1]
    print(f"Login incorrecto, le quedan {3-intentos_realizados} intento(s).")
```

7. Escribir una función que reciba una frase y devuelva un diccionario con las palabras que contiene y su longitud.

```python
def frase_a_diccionario(frase):
  palabras=frase.split()
  diccionario={}
  for palabra in palabras:
    diccionario[palabra]=len(palabra)
  return diccionario

frase_a_diccionario("Hola qué tal cómo te va")
```

8. Realiza una función separar(lista) que tome una lista de números enteros y devuelva dos listas ordenadas. La primera con los números pares y la segunda con los números impares.

```python
def separar(lista):
  lista_pares=[]
  lista_impares=[]
  for numero in lista:
    if numero % 2 == 0:
      lista_pares.append(numero)
    else:
      lista_impares.append(numero)
  lista_pares.sort()
  lista_impares.sort()
  return lista_pares, lista_impares

separar([4,10,3,7,8,5])
```

9. Definir una función que calcule la longitud de una lista o una cadena dada. No utilizar la función len().

```python
def longitud(lista):
  contador = 0
  for i in lista:
    contador += 1
  return contador

print(longitud([1,2,3,4]))
print(longitud("Alberto"))
```

10. Definir una función inversa() que calcule la inversión de una cadena. Por ejemplo la cadena
"estoy probando" debería devolver la cadena "odnaborp yotse"

```python
def inversa(cadena):
  lista=list(cadena)
  lista.reverse()
  cadena_inversa="".join(lista)
  return cadena_inversa

inversa("estoy probando")
```
