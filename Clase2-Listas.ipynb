15. Escribir un programa que almacene los vectores (1,2,3) y (-1,0,2) en dos listas y muestre por pantalla su producto escalar.

16. Escribir un programa que almacene dos matrices en una lista y muestre por pantalla su producto. Nota: Para representar matrices mediante listas usar listas anidadas, representando cada vector fila en una lista.

```python
mat1=[[1,2],[3,4]]
mat2=[[0,-1],[-2,5]]

prod=[]
for i in range(len(mat1[0])):
    fila=[]
    for j in range(len(mat1[0])):
        elem=0
        for k in range(len(mat1[0])):
            elem+=mat1[i][k]*mat2[k][j]
        fila.append(elem)
    prod.append(fila)
print(prod)
```
