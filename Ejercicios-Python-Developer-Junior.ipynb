16. Escribir un programa que almacene dos matrices en una lista y muestre por pantalla su producto. Nota: Para representar matrices mediante listas usar listas anidadas, representando cada vector fila en una lista.

```python
mat1=[[1,2],[3,4]]
mat2=[[0,-1],[-2,5]]

prod=[]
for i in range(len(mat1[0])):
    fila=[]
    for j in range(len(mat1[0])):
        elem=0
        for k in range(len(mat1[0])):
            elem+=mat1[i][k]*mat2[k][j]
        fila.append(elem)
    prod.append(fila)
print(prod)
```

```python
tupla=(10,17,11,23,16)

lista_primos=[]
for i in tupla:
  bandera=0
  for n in range(2,i):
    if i % n == 0:
      bandera=1
  if bandera==0:
    lista_primos.append(i)
tupla_primos=tuple(lista_primos)
print(f"a) Los primos son {tupla_primos}.")

sumatoria=0
for i in tupla:
  sumatoria+=i
promedio=sumatoria/len(tupla)
print(f"b) La sumatoria es {sumatoria} y el promedio es {promedio}.")

lista_factoriales=[]
for i in tupla:
  factorial=1
  for j in range(2,i+1):
    factorial=factorial*j
  lista_factoriales.append(factorial)
tupla_factoriales=tuple(lista_factoriales)
print(f"c) Los factoriales son {tupla_factoriales}.")
```

```python
def separar(lista):
  lista_pares=[]
  lista_impares=[]
  for numero in lista:
    if numero % 2 == 0:
      lista_pares.append(numero)
    else:
      lista_impares.append(numero)
  lista_pares.sort()
  lista_impares.sort()
  return lista_pares, lista_impares

separar([4,10,3,7,8,5])
```

```python
def comparar(lista,k):
  menores=[]
  iguales=[]
  mayores=[]
  for i in lista:
    if i < k:
      menores.append(i)
    elif i == k:
      iguales.append(i)
    else:
      mayores.append(i)
  return menores,iguales,mayores

def multiplos(lista,k):
  mult=[]
  for i in lista:
    if i % k == 0:
      mult.append(i)
  return mult

N=int(input("Ingrese la cantidad de n√∫meros: "))
lista=[]
for i in range(N):
  elemento=int(input(f"Ingrese el {i+1}¬∞ elemento: "))
  lista.append(elemento)
k=int(input("Ingrese el valor de k: "))

a,b,c=comparar(lista,k)
print(f"Los n√∫meros menores a k son: {a}, los iguales a k son: {b} y los mayores a k son: {c}.")

m=multiplos(lista,k)
print(f"Los m√∫ltiplos de k son: {m}.")
```

```python
class Fraccion(object):
  def __init__(self, numerador, denominador):
    self.numerador = numerador
    self.denominador = denominador

  def __add__(self, otro):
    if self.denominador == otro.denominador:
      return Fraccion(self.numerador + otro.numerador, self.denominador)
    else:
      return "Las fracciones no tienen igual denominador."

  def __str__(self):
    return str(self.numerador) + "/" + str(self.denominador)

x = Fraccion(3,5)
y = Fraccion(6,5)
print(x+y)

x = Fraccion(3,5)
y = Fraccion(6,7)
print(x+y)
```
