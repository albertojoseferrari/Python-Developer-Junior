```python
mat1=[[1,2],[3,4]]
mat2=[[0,-1],[-2,5]]

prod=[]
for i in range(len(mat1[0])):
    fila=[]
    for j in range(len(mat1[0])):
        elem=0
        for k in range(len(mat1[0])):
            elem+=mat1[i][k]*mat2[k][j]
        fila.append(elem)
    prod.append(fila)
print(prod)
```

```python
stock={}
opcion="0"
venta_total=[]
while opcion!="f" and opcion!="F":
  print()
  print("MENÚ:")
  print("1) Cargar un producto")
  print("2) Vender un producto")
  print("3) Listar cantidad de productos con stock mínimo debajo de N")
  print("4) Calcular el monto total vendido en el día")
  print("5) Baja de un producto")
  print("f o F) Fin del programa")
  print()
  opcion=input("Ingrese una opción del menú: ")
  print()
  if opcion=="1":
    producto=input("Ingrese el producto: ")
    peso=float(input(f"Ingrese el peso unitario del producto {producto}: "))
    valido=False
    if peso > 0:
      valido=True
    while not valido:
      print("El peso ingresado no es válido.")
      peso=float(input(f"Ingrese un nuevo peso unitario del producto {producto}: "))
      if peso > 0:
        valido=True
    cantidad=int(input(f"Ingrese la cantidad del producto {producto}: "))
    valido=False
    if cantidad > 0:
      valido=True
    while not valido:
      print("La cantidad ingresada no es válido.")
      cantidad=int(input(f"Ingrese una nueva cantidad del producto {producto}: "))
      if cantidad > 0:
        valido=True
    precio=float(input(f"Ingrese el precio del producto {producto}:"))
    valido=False
    if precio > 0:
      valido=True
    while not valido:
      print("El precio ingresado no es válido.")
      precio=float(input(f"Ingrese un nuevo precio del producto {producto}:"))
      if precio > 0:
        valido=True
    stock[producto]=[peso,cantidad,precio]
  if opcion=="2":
    producto=input("Ingrese un producto: ")
    valido=False
    for i in stock:
      if producto==i:
        valido=True
    if not valido:
      print("El producto ingresado no existe.")
    else:
      cantidad_vendida=int(input(f"Ingrese la cantidad vendida del producto {producto}: "))
      valido=False
      if stock[producto][1] > cantidad_vendida:
        valido=True
      if not valido:
        print("La cantidad vendida excede el stock del producto.")
      else:
        stock[producto][1] -= cantidad_vendida
        venta=cantidad_vendida*stock[producto][2]
        print(f"El total de la venta es de ${venta}.")
        venta_total.append(venta)
  if opcion=="3":
    N=int(input("Ingrese un valor N: "))
    valido=False
    if N > 0:
      valido=True
    while not valido:
      print("El N ingresado no es válido.")
      N=int(input("Ingrese un nuevo valor de N: "))
      if N > 0:
        valido=True
    for i in stock:
      if stock[i][1]<N:
        print(f"El producto {i} tiene cantidad de stock {stock[i][1]} que es inferior a N.")
  if opcion=="4":
    suma=0
    for i in range(len(venta_total)):
      suma+=venta_total[i]
    print(f"La cantidad de ventas realizadas es {len(venta_total)} y el total de todas las ventas es ${suma}.")
  if opcion=="5":
    producto_a_eliminar=input("Ingrese el producto a eliminar: ")
    valido=False
    for i in stock:
      if producto_a_eliminar==i:
        valido=True
    if valido:
      stock.pop(producto_a_eliminar)
      print("Se eliminó el producto.")
    else:
      print("No existe el producto a eliminar.")
```

```python
def separar(lista):
  lista_pares=[]
  lista_impares=[]
  for numero in lista:
    if numero % 2 == 0:
      lista_pares.append(numero)
    else:
      lista_impares.append(numero)
  lista_pares.sort()
  lista_impares.sort()
  return lista_pares, lista_impares

separar([4,10,3,7,8,5])
```

*   Devuelva tres listas, una con los menores, otra con los mayores y otra con los valores iguales a k.
*   Devuelva una lista con aquellos que son múltiplos de k.

```python
def comparar(lista,k):
  menores=[]
  iguales=[]
  mayores=[]
  for i in lista:
    if i < k:
      menores.append(i)
    elif i == k:
      iguales.append(i)
    else:
      mayores.append(i)
  return menores,iguales,mayores

def multiplos(lista,k):
  mult=[]
  for i in lista:
    if i % k == 0:
      mult.append(i)
  return mult

N=int(input("Ingrese la cantidad de números: "))
lista=[]
for i in range(N):
  elemento=int(input(f"Ingrese el {i+1}° elemento: "))
  lista.append(elemento)
k=int(input("Ingrese el valor de k: "))

a,b,c=comparar(lista,k)
print(f"Los números menores a k son: {a}, los iguales a k son: {b} y los mayores a k son: {c}.")

m=multiplos(lista,k)
print(f"Los múltiplos de k son: {m}.")
```

```python
base_de_datos={}

contador_D = 0
monto_total_D = 0

contador_E = 0
monto_total_E = 0

def carga():
  clave_ingresada = int(input("Ingrese un número de cliente: "))
  if clave_ingresada in base_de_datos:
    print(f"El apellido del cliente es {base_de_datos[clave_ingresada][0]} y su saldo es {base_de_datos[clave_ingresada][1]}.")
  else:
    respuesta = input("El cliente no existe, ¿desea agregarlo? S/N: ").upper()
    if respuesta == "S":
      apellido_ingresado = input("Ingrese el apellido del cliente: ")
      base_de_datos[clave_ingresada]=[apellido_ingresado,0]

def operacion():
  clave_ingresada = int(input("Ingrese un número de cliente: "))
  if clave_ingresada in base_de_datos:
    operacion = input("¿Desea realizar un depósito (D) o una extracción (E)?: ").upper()
    if operacion == "D":
      global contador_D
      contador_D += 1
      monto = float(input("Ingrese el monto a depositar: "))
      global monto_total_D
      monto_total_D += monto
      base_de_datos[clave_ingresada][1] += monto
    elif operacion == "E":
      global contador_E
      contador_E += 1
      monto = float(input("Ingrese el monto a extraer: "))
      global monto_total_E
      monto_total_E += monto
      base_de_datos[clave_ingresada][1] -= monto
    else:
      print("La operación ingresada no es correcta.")
  else:
    print("El cliente no existe.")

def registro():
  global contador_D
  global monto_total_D
  global contador_E
  global monto_total_E
  print(f"Se hicieron un total de {contador_D} depósitos por un monto total de {monto_total_D}.")
  print(f"Se hicieron un total de {contador_E} extracciones por un monto total de {monto_total_E}.")

while True:

  print('''
        MENÚ:

        1. Cargar cliente
        2. Operación
        3. Registro de operaciones
        S-s. Salir

        ''')

  opcion = input("Elija una opción: ").upper()

  if opcion == "1":
    carga()

  elif opcion == "2":
    operacion()

  elif opcion == "3":
    registro()

  elif opcion == "S":
    print("Usted eligió Salir.")
    break

  else:
    print("La opción ingresada no es correcta.")
```

```python
class Fraccion(object):
  def __init__(self, numerador, denominador):
    self.numerador = numerador
    self.denominador = denominador

  def __add__(self, otro):
    if self.denominador == otro.denominador:
      return Fraccion(self.numerador + otro.numerador, self.denominador)
    else:
      return "Las fracciones no tienen igual denominador."

  def __str__(self):
    return str(self.numerador) + "/" + str(self.denominador)

x = Fraccion(3,5)
y = Fraccion(6,5)
print(x+y)

x = Fraccion(3,5)
y = Fraccion(6,7)
print(x+y)
```
