6. Es sabido que un tri√°ngulo se forma con tres segmentos o lados de una determinada longitud pero, dados 3 segmentos de longitud cualesquiera, no siempre es posible construir un tri√°ngulo. Implemente un c√≥digo que permita validar si a partir de 3 segmentos A, B y C los mismos representan los lados de un tri√°ngulo v√°lido. Luego, debe informar qu√© tipo de tri√°ngulo es.

```python
A=float(input ("ingrese el lado A "))
B=float(input ("ingrese el lado B "))
C=float(input ("ingrese el lado C "))
if A+B>C and A+C>B and B+C>A:
   if A==B and B==C:
      print("Es un tri√°ngulo equil√°tero")
   elif A==B or A==C or B==C:
      print("Es un tri√°ngulo is√≥sceles no equil√°tero")
   else:
      print("Es un tri√°ngulo escaleno")
else:
   print("No es un tri√°ngulo")
```

7. Escribir un programa que permita
- ingresar un n√∫mero N (entero positivo, si se ingresa otra cosa, se deber√° mostrar un mensaje indicando un error)
- ingresar un n√∫mero X (real positivo que debe estar entre 0 y œÄ , si se ingresa otra cosa, se deber√° mostrar un mensaje indicando un error)
- luego calcular y mostrar la aproximaci√≥n al Seno del √°ngulo X como Sen (X) = (ùë• - ùë•3 / 3 ! + x5 / 5 ! - x7 / 7 ! + ‚Ä¶) NOTA: N expresa la cantidad de t√©rminos a considerar en la serie de Taylor, utilizada para aproximar el Seno de X. Verificar que... a mayor N se obtiene una mejor aproximaci√≥n.

```python
PI=3.14159
N=int(input("ingrese un n√∫mero N entero positivo: "))
if N <= 0:
    print("N no es positivo")
X=float(input("ingrese un n√∫mero X real positivo entre 0 y œÄ: "))
if X<0 or X>PI:
    print("X no est√° entre 0 y œÄ")
seno=X
factorial=1
for i in range(1,N):
    factorial=factorial*(2*i)*(2*i+1)
    seno=seno+(-1)**(i)*X**(2*i+1)/factorial
print(seno)
```

8. Realizar un programa que permita adivinar un n√∫mero. Para ello se solicitara el ingreso por teclado del numero a adivinar (entero). Luego se ir√°n solicitando mas n√∫meros y se deber√° ir averiguando si el n√∫mero a adivinar es mayor o menor que el introducido (ir mostrando carteles indicativos). El programa termina cuando se acierta el n√∫mero.

```python
num=int(input("ingrese el n√∫mero a adivinar: "))
intento=int(input("intente adivinar el n√∫mero: "))
while intento != num:
    if intento < num:
        print("El n√∫mero a adivinar es m√°s grande")
        intento=int(input("intente adivinar el n√∫mero: "))
    else:
        print("El n√∫mero a adivinar es m√°s chico")
        intento=int(input("intente adivinar el n√∫mero: "))
print("¬°Adivinaste!")
```
