1. Dada una lista de números enteros y un entero k, escribir una función que:
*   Devuelva tres listas, una con los menores, otra con los mayores y otra con los valores iguales a k.
*   Devuelva una lista con aquellos que son múltiplos de k.

```python
def comparar(lista,k):
  menores=[]
  iguales=[]
  mayores=[]
  for i in lista:
    if i < k:
      menores.append(i)
    elif i == k:
      iguales.append(i)
    else:
      mayores.append(i)
  return menores,iguales,mayores

def multiplos(lista,k):
  mult=[]
  for i in lista:
    if i % k == 0:
      mult.append(i)
  return mult

N=int(input("Ingrese la cantidad de números: "))
lista=[]
for i in range(N):
  elemento=int(input(f"Ingrese el {i+1}° elemento: "))
  lista.append(elemento)
k=int(input("Ingrese el valor de k: "))

a,b,c=comparar(lista,k)
print(f"Los números menores a k son: {a}, los iguales a k son: {b} y los mayores a k son: {c}.")

m=multiplos(lista,k)
print(f"Los múltiplos de k son: {m}.")
```

3. Crear un programa que presente mediante una función el siguiente menú:

1) Cargar lista

2) Mostrar lista

3) Ordenar lista

S) Salir

No se puede ingresar al punto 2 o 3 sin haber ingresado previamente al 1.
Solo se puede salir ingresando S.

Al seleccionar la opción 1: Se llama a la función carga que realiza lo siguiente:

Si es la primera vez, se cargan valores numéricos a la lista.

Después de la primera vez, se agregan valores numéricos a la lista previamente cargada (no se sobrescriben los valores que se cargaron la primera vez)

Se debe cargar un elemento y preguntar al usuario si desea cargar otro. Si ingresa S se carga un elemento más, si ingresa N se sale de la función.

Al seleccionar la opción 2: Se muestra la lista en pantalla mediante una función.

Al seleccionar la opción 3: Se llama a una función pasándole como argumento M o m.

Si se le pasa "M" se debe ordenar la lista de Mayor a menor

Si se le pasa "m" se debe ordenar la lista de menor a Mayor.


```python
def menu():
  print('''
        MENÚ:

        1) Cargar lista
        2) Mostrar lista
        3) Ordenar lista
        S) Salir.

        ''')

lista = []

def carga(lista):
    respuesta = "S"
    while respuesta.upper() == "S":
      elemento = float(input("Ingrese un elemento: "))
      lista.append(elemento)
      respuesta = input("¿Desea cargar otro elemento? S/N: ")

def muestra(lista):
  print(f"La lista es: {lista}.")

def ordenar(lista,orden):
  if orden == "M":
    lista.sort(reverse=True)
  else:
    lista.sort()

bandera = 0

while True:

  menu()

  opcion = input("Elija una opción: ").upper()

  if opcion == "1":
    bandera = 1
    carga(lista)

  elif opcion == "2":
    if bandera == 0:
      print("No se puede ingresar al punto 2 sin haber ingresado previamente al 1.")
    else:
      muestra(lista)

  elif opcion == "3":
    if bandera == 0:
      print("No se puede ingresar al punto 3 sin haber ingresado previamente al 1.")
    else:
      orden_elegido = input("Ingrese M si desea ordenar la lista de mayor a menor o m si desea ordenarla de menor a mayor: ")
      ordenar(lista,orden_elegido)

  elif opcion == "S":
    print("Usted eligió Salir.")
    break

  else:
    print("La opción ingresada no es correcta.")
```

4. Se desea gestionar las cuentas de los clientes de un Banco.

Para ello se solicita un programa que deberá crear un diccionario con el número de cliente, (clave) y el Apellido y saldo del mismo.

El programa deberá presentar el siguiente menú:

1.	Cargar cliente

2.	Operación

3.	Registro de operaciones

S-s. Salir

Si elige el punto 1, deberá ingresar por teclado el número de cliente (validar) si existe, deberá mostrar los datos del  mismo. Si no existe se preguntará si se desea agregar. Si la repuesta es si, agregar al diccionario los datos con saldo 0.

Si elige el punto 2 deberá solicitar por teclado un número de cliente (si existe) y un código de operación. Si es “D” (deposito) permitirá ingresar un valor que es el monto a depositar. Si el código “E” (extracción) permitirá descontar del saldo ese importe. Si se ingresa un nro. de cliente no valido mostrar el cartel correspondiente.

Si elige el punto 3 deberá mostrar la cantidad de operaciones D y E que se hicieron y el monto total de las mismas.

El programa finaliza cuando elige s o S.

Cada uno de los puntos 1,2 y 3 deberá hacerlo con funciones.


```python
base_de_datos={}

contador_D = 0
monto_total_D = 0

contador_E = 0
monto_total_E = 0

def carga():
  clave_ingresada = int(input("Ingrese un número de cliente: "))
  if clave_ingresada in base_de_datos:
    print(f"El apellido del cliente es {base_de_datos[clave_ingresada][0]} y su saldo es {base_de_datos[clave_ingresada][1]}.")
  else:
    respuesta = input("El cliente no existe, ¿desea agregarlo? S/N: ").upper()
    if respuesta == "S":
      apellido_ingresado = input("Ingrese el apellido del cliente: ")
      base_de_datos[clave_ingresada]=[apellido_ingresado,0]

def operacion():
  clave_ingresada = int(input("Ingrese un número de cliente: "))
  if clave_ingresada in base_de_datos:
    operacion = input("¿Desea realizar un depósito (D) o una extracción (E)?: ").upper()
    if operacion == "D":
      global contador_D
      contador_D += 1
      monto = float(input("Ingrese el monto a depositar: "))
      global monto_total_D
      monto_total_D += monto
      base_de_datos[clave_ingresada][1] += monto
    elif operacion == "E":
      global contador_E
      contador_E += 1
      monto = float(input("Ingrese el monto a extraer: "))
      global monto_total_E
      monto_total_E += monto
      base_de_datos[clave_ingresada][1] -= monto
    else:
      print("La operación ingresada no es correcta.")
  else:
    print("El cliente no existe.")

def registro():
  global contador_D
  global monto_total_D
  global contador_E
  global monto_total_E
  print(f"Se hicieron un total de {contador_D} depósitos por un monto total de {monto_total_D}.")
  print(f"Se hicieron un total de {contador_E} extracciones por un monto total de {monto_total_E}.")

while True:

  print('''
        MENÚ:

        1. Cargar cliente
        2. Operación
        3. Registro de operaciones
        S-s. Salir

        ''')

  opcion = input("Elija una opción: ").upper()

  if opcion == "1":
    carga()

  elif opcion == "2":
    operacion()

  elif opcion == "3":
    registro()

  elif opcion == "S":
    print("Usted eligió Salir.")
    break

  else:
    print("La opción ingresada no es correcta.")
```
