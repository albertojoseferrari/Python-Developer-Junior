# CÓDIGO

Una empresa desea almacenar datos de Nro de empleado, Nombre, Edad y Sueldo  de sus empleados.

Elaborar un programa en Python que muestre el siguiente menú:

1) Carga de valores del empleado

2) Cálculo de cantidad de empleados con edad mayor a X

3) Búsqueda del mayor sueldo

4) Listado de los registros almacenados en el diccionario (un par clave:valor por línea)

S o s) Salir del programa.

Si elige el punto 1:
Cada vez que elige el punto 1 deberá cargar en un diccionario los datos del empleado: Numero del empleado (clave) y Nombre, Edad, Sueldo (valores)
Validar que el nro del empleado que se ingrese sea positivo, de lo contrario, reingresar. Ayuda: utilizar while para el reingreso.
Ayuda: el diccionario debería tener como clave el nro del empleado y como datos una lista por cada empleado. Por ejemplo:

Datos={2:[Juan, Perez, 2500], 4:[Gonzalez, Maria, 3000],5:[Sanchez, Jose, 6500]}

Si elige el punto 2:

Se deberá ingresar por teclado en el Programa Principal un numero X (validar que sea mayor a 25) y luego llamar a una función de nombre Calculo la cual deberá devolver la cantidad de empleados cuya edad es mayor al valor X ingresado.

Ayuda: : fijarse que datos necesita la función para poder hacer este calculo

Si elige el punto 3:

Llamar a una función de nombre Búsqueda la cual deberá buscar y mostrar en la función el empleado que tenga mayor sueldo.

Si elige el punto 4:

Deberá mostrar con una función el diccionario completo.

Si elige S o S sale del programa.

```python
# funciones a utilizar

def Calculo(valor):

  contador = 0

  for empleado in Datos:
    if Datos[empleado][1] > valor:
      contador += 1

  return print(f"La cantidad de empleados con edad mayor a {valor} es {contador}.")

def Busqueda():

  maximo_sueldo = 0

  for empleado in Datos:
    if Datos[empleado][2] > maximo_sueldo:
      maximo_sueldo = Datos[empleado][2]
      empleado_maximo_sueldo = Datos[empleado][0]

  return print(f"El empleado de mayor sueldo es {empleado_maximo_sueldo} y su sueldo es {maximo_sueldo}.")

def Muestra():

  for clave, valor in Datos.items():
    print(f"El empleado número {clave} es {valor[0]}, su edad es {valor[1]} y su sueldo es {valor[2]}.")

# programa principal

Datos={}

while True:

  print('''
        MENÚ:

        1) Carga de valores del empleado
        2) Cálculo de cantidad de empleados con edad mayor a X
        3) Búsqueda del mayor sueldo
        4) Listado de los registros almacenados en el diccionario
        S o s) Salir del programa

        ''')

  opcion = input("Elija una opción: ").upper()

  if opcion == "1":

    clave_ingresada = int(input("Ingrese el número de empleado: "))

    # validación de la clave
    while clave_ingresada <= 0:
      print("El número ingresado debe ser positivo.")
      clave_ingresada = int(input("Ingrese nuevamente el número de empleado: "))

    nombre_ingresado = input("Ingrese el nombre del empleado: ")
    edad_ingresada = int(input("Ingrese la edad del empleado: "))
    sueldo_ingresado = float(input("Ingrese el sueldo del empleado: "))

    Datos[clave_ingresada]=[nombre_ingresado,edad_ingresada,sueldo_ingresado]

  elif opcion == "2":

    X = int(input("Ingrese el valor de X: "))

    # validación de X
    while X <= 24:
      print("El valor de X debe ser al menos 25.")
      X = int(input("Ingrese nuevamente el valor de X: "))

    Calculo(X)

  elif opcion == "3":

    Busqueda()

  elif opcion == "4":

    Muestra()

  elif opcion == "S":
    print("Usted eligió Salir.")
    break

  else:
    print("La opción ingresada no es correcta.")
```
