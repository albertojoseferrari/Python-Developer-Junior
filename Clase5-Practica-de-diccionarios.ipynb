1. Realizar un programa que permita la carga de datos de estudiantes. Los datos a cargar son: Nombre Completo, Nota 1 y Nota 2

Se debe mostrar el siguiente Menu:

    1- Carga de estudiante
    2- Mostrar la cantidad de estudiantes cargados
    3- Mostrar el promedio de cada estudiante
    4- Mostrar promedio mas alto
    S- Salir

Si **ingresa 1**: Se deberá cargar un DICCIONARIO que contenga como CLAVE el nombre completo del estudiante EN MAYUSCULAS, y como valor, una lista que contenga sus dos notas de evaluaciones.

Se debera cargar un estudiante por vez, es decir,
luego de cada carga se debera mostrar el menu nuevamente.

Si ya se cargo el estudiante, mostrar un cartel ESTUDIANTE YA CARGADO

Si **ingresa 2**: Mostrará por pantalla la cantidad de estudiantes cargados.

Si **ingresa 3:** Listar los estudiantes con sus promedios.

Si **ingresa 4:** Mostrar el promedio mas alto.

Para salir debe ingresar S o s

```python
diccio={}
opcion="0"
while opcion!="S" and opcion!="s":
  print("MENÚ:")
  print("1- Carga de estudiante")
  print("2- Mostrar la cantidad de estudiantes cargados")
  print("3- Mostrar el promedio de cada estudiante")
  print("4- Mostrar promedio mas alto")
  print("S- Salir")
  opcion=input("Ingrese una opción: ")
  if opcion=="1":
    bandera=0
    nombre=input("Ingrese el nombre completo del estudiante: ")
    nombre=nombre.upper()
    for i in diccio:
      if i==nombre:
        print("El estudiante ya se ha cargado.")
        bandera=1
    if bandera==0:
      nota1=int(input("Ingrese la nota 1 del estudiante: "))
      nota2=int(input("Ingrese la nota 2 del estudiante: "))
      diccio[nombre]=[nota1,nota2]
  if opcion=="2":
    print("La cantidad de estudiantes cargados es", len(diccio))
  if opcion=="3":
    estudiantes=[]
    promedios=[]
    for i in diccio:
      estudiantes.append(i)
      promedios.append((diccio[i][0]+diccio[i][1])/2)
    lista=[]
    for item in zip(estudiantes,promedios):
      lista.append(item)
    print("El promedio de cada estudiante es:", lista)
  if opcion=="4":
    promedios=[]
    for i in diccio:
      promedios.append((diccio[i][0]+diccio[i][1])/2)
    print("El promedio más alto es", max(promedios))
```

2. DICCIONARIOS CON LISTAS - Crear un programa que permita guardar el stock de un conjunto de productos de un negocio.

Los datos a almacenar son el nombre del producto (clave) y para cada producto tenemos los siguientes valores: peso unitario del producto (valor real), cantidad (valor entero), precio del producto (valor real).

Se debe presentar un menú

1) Cargar un producto

2) Vender un producto

3) Listar cantidad de productos con stock mínimo debajo de N

4) Calcular el monto total vendido en el día

5) Baja de un producto

f o F) Fin del programa

Si elige 1, se deberá cargar un producto al diccionario. Se carga la clave y la lista con los valores de peso, cantidad y precio. Si los valores ingresados son negativos o cero se debe solicitar reingreso SOLO del valor incorrecto. Ver que si la clave ya existe en el diccionario puede modificar los valores.


Si elige 2, debe ingresar un nombre de producto y la cantidad a vender de ese producto, luego debe descontar la cantidad vendida del stock y mostrar el total de la venta. Si no se puede vender, por que NO hay stock suficiente o porque el producto no existe se debe mostrar un cartel. Las ventas del día (total) se deben ir guardando en una lista que se utilizará en el punto 4.


Si elige 3, debe ingresar un valor N, entero positivo (validar sino reingresar), y mostrar los productos que tengan cantidad de stock inferior a N.


Si elige 4, utiliza la lista que contiene los totales de cada venta y muestra la cantidad de ventas realizadas y total de todas las ventas.


Si elige 5, debe ingresar un nombre de producto y, si existe la clave elimina el producto y muestra un cartel: "Se eliminó el producto". Si no existe la clave se muestra un cartel "No existe el producto a eliminar".

```python
stock={}
opcion="0"
venta_total=[]
while opcion!="f" and opcion!="F":
  print()
  print("MENÚ:")
  print("1) Cargar un producto")
  print("2) Vender un producto")
  print("3) Listar cantidad de productos con stock mínimo debajo de N")
  print("4) Calcular el monto total vendido en el día")
  print("5) Baja de un producto")
  print("f o F) Fin del programa")
  print()
  opcion=input("Ingrese una opción del menú: ")
  print()
  if opcion=="1":
    producto=input("Ingrese el producto: ")
    peso=float(input(f"Ingrese el peso unitario del producto {producto}: "))
    valido=False
    if peso > 0:
      valido=True
    while not valido:
      print("El peso ingresado no es válido.")
      peso=float(input(f"Ingrese un nuevo peso unitario del producto {producto}: "))
      if peso > 0:
        valido=True
    cantidad=int(input(f"Ingrese la cantidad del producto {producto}: "))
    valido=False
    if cantidad > 0:
      valido=True
    while not valido:
      print("La cantidad ingresada no es válido.")
      cantidad=int(input(f"Ingrese una nueva cantidad del producto {producto}: "))
      if cantidad > 0:
        valido=True
    precio=float(input(f"Ingrese el precio del producto {producto}:"))
    valido=False
    if precio > 0:
      valido=True
    while not valido:
      print("El precio ingresado no es válido.")
      precio=float(input(f"Ingrese un nuevo precio del producto {producto}:"))
      if precio > 0:
        valido=True
    stock[producto]=[peso,cantidad,precio]
  if opcion=="2":
    producto=input("Ingrese un producto: ")
    valido=False
    for i in stock:
      if producto==i:
        valido=True
    if not valido:
      print("El producto ingresado no existe.")
    else:
      cantidad_vendida=int(input(f"Ingrese la cantidad vendida del producto {producto}: "))
      valido=False
      if stock[producto][1] > cantidad_vendida:
        valido=True
      if not valido:
        print("La cantidad vendida excede el stock del producto.")
      else:
        stock[producto][1] -= cantidad_vendida
        venta=cantidad_vendida*stock[producto][2]
        print(f"El total de la venta es de ${venta}.")
        venta_total.append(venta)
  if opcion=="3":
    N=int(input("Ingrese un valor N: "))
    valido=False
    if N > 0:
      valido=True
    while not valido:
      print("El N ingresado no es válido.")
      N=int(input("Ingrese un nuevo valor de N: "))
      if N > 0:
        valido=True
    for i in stock:
      if stock[i][1]<N:
        print(f"El producto {i} tiene cantidad de stock {stock[i][1]} que es inferior a N.")
  if opcion=="4":
    suma=0
    for i in range(len(venta_total)):
      suma+=venta_total[i]
    print(f"La cantidad de ventas realizadas es {len(venta_total)} y el total de todas las ventas es ${suma}.")
  if opcion=="5":
    producto_a_eliminar=input("Ingrese el producto a eliminar: ")
    valido=False
    for i in stock:
      if producto_a_eliminar==i:
        valido=True
    if valido:
      stock.pop(producto_a_eliminar)
      print("Se eliminó el producto.")
    else:
      print("No existe el producto a eliminar.")
```

3. DICCIONARIOS CON DICCIONARIOS - Escribir un programa que permita gestionar la base de datos de clientes de una empresa.

Los clientes se guardarán en un diccionario en el que la clave de cada cliente será su DNI, y el valor será otro diccionario con los datos del cliente (nombre, dirección, teléfono, correo, tipocliente), donde preferente tendrá el valor True si se trata de un cliente preferente.

El programa debe preguntar al usuario por una opción del siguiente menú:

(1) Añadir cliente,

(2) Eliminar cliente,

(3) Mostrar cliente,

(4) Listar todos los clientes,

(5) Listar clientes preferentes,

(6) Terminar.

En función de la opción elegida el programa tendrá que hacer lo siguiente:

1. Preguntar los datos del cliente, crear un diccionario con los datos y añadirlo a la base de datos.
2. Preguntar por el DNI del cliente y eliminar sus datos de la base de datos.
3. Preguntar por el DNI del cliente y mostrar sus datos.
4. Mostrar lista de todos los clientes de la base datos con su DNI y nombre.
5. Mostrar la lista de clientes preferentes de la base de datos con su DNI y nombre.
6. Terminar el programa.

```python
base_de_datos={}
opcion="0"
while opcion != "6":
  print()
  print("MENÚ:")
  print("(1) Añadir cliente")
  print("(2) Eliminar cliente")
  print("(3) Mostrar cliente")
  print("(4) Listar todos los clientes")
  print("(5) Listar clientes preferentes")
  print("(6) Terminar")
  print()
  opcion=input("Ingrese una opción del menú: ")
  print()
  if opcion=="1":
    datos={}
    DNI=int(input("Ingrese el DNI del cliente: "))
    nombre=input("Ingrese el nombre del cliente: ")
    datos["nombre"]=nombre
    direccion=input("Ingrese la dirección del cliente: ")
    datos["direccion"]=direccion
    telefono=int(input("Ingrese el teléfono del cliente: "))
    datos["telefono"]=telefono
    correo=input("Ingrese el correo del cliente: ")
    datos["correo"]=correo
    preferente=input("¿Se trata de un cliente preferente? s/n: ")
    if preferente == "s":
      tipocliente = True
    else:
      tipocliente = False
    datos["tipocliente"]=tipocliente
    base_de_datos[DNI]=datos
  if opcion=="2":
    DNI_a_eliminar=int(input("Ingrese el DNI del cliente a eliminar: "))
    valido=False
    for i in base_de_datos:
      if DNI_a_eliminar==i:
        valido=True
    if valido:
      base_de_datos.pop(DNI_a_eliminar)
      print("Se eliminó el cliente.")
    else:
      print("El DNI del cliente a eliminar no está cargado en la base de datos.")
  if opcion=="3":
    DNI_a_mostrar=int(input("Ingrese el DNI del cliente a mostrar: "))
    valido=False
    for i in base_de_datos:
      if DNI_a_mostrar==i:
        valido=True
    if valido:
      print(base_de_datos[DNI_a_mostrar])
    else:
      print("El DNI del cliente a mostrar no está cargado en la base de datos.")
  if opcion=="4":
    lista=[]
    DNIs=[]
    nombres=[]
    for i in base_de_datos:
      DNIs.append(i)
      nombres.append(base_de_datos[i]["nombre"])
    for item in zip(DNIs,nombres):
      lista.append(item)
    print("Lista de todos los clientes de la base de datos:", lista)
  if opcion=="5":
    lista=[]
    DNIs=[]
    nombres=[]
    for i in base_de_datos:
      if base_de_datos[i]["tipocliente"]:
        DNIs.append(i)
        nombres.append(base_de_datos[i]["nombre"])
    for item in zip(DNIs,nombres):
      lista.append(item)
    print("Lista de clientes preferentes de la base de datos:", lista)
```

4. LISTAS CON DICCIONARIOS - Se desea desarrollar un programa para el control de empleados en una empresa. Los empleados deben almacenar en una lista.

Debe mostrar en pantalla el siguiente menú:


MENU:


1: Carga de empleados


2: Cantidad de empleados por turno


3: Promedio de horas por sector


S: Salir

Si el usuario elige la **opcion 1**, debe preguntar cuántos empleados desea cargar (mayor a 0 y menor a 1000).


Por cada empleado se debe crear un diccionario con las siguientes claves y valores:


“legajo” : un número entero entre 0 y 1000


“edad” : un número entero entre 18 y 65


“sector” : un caracter (M: Mantenimiento, A: Administración, O: Operarios)


“turno” : un caracter (M o T)


“horas”: un número entre 10 y 40


Una vez cargado, el diccionario para un empleado se debe agregar a la lista de empleados

Si el usuario elige la **opcion 2** , debe ingresar un turno elegido por el usuario y mostrar en pantalla la cantidad de empleados que trabajan en dicho turno.



Si el usuario elige la **opcion 3** , debe ingresar un sector elegido por el usuario y el programa debe calcular y mostrar el promedio de horas de trabajo en el sector elegido.

```python
lista_de_empleados=[]
opcion="0"
while opcion != "S":
  print()
  print("MENÚ:")
  print("1: Carga de empleados")
  print("2: Cantidad de empleados por turno")
  print("3: Promedio de horas por sector")
  print("S: Salir")
  print()
  opcion=input("Ingrese una opción del menú: ")
  opcion=opcion.upper()
  print()
  if opcion=="1":
    N=int(input("¿Cuántos empleados desea cargar? "))
    valido=False
    if N > 0 and N < 1000:
      valido=True
    while not valido:
      print("La cantidad ingresada debe estar entre 0 y 1000.")
      N=int(input("¿Cuántos empleados desea cargar? "))
      if N > 0 and N < 1000:
        valido=True
    for i in range(N):
      datos={}
      legajo=int(input("Ingrese el legajo del empleado: "))
      valido=False
      if legajo > 0 and legajo < 1000:
        valido=True
      while not valido:
        print("El legajo ingresado debe estar entre 0 y 1000.")
        legajo=int(input("Ingrese el legajo del empleado: "))
        if legajo > 0 and legajo < 1000:
          valido=True
      datos["legajo"]=legajo
      edad=int(input("Ingrese la edad del empleado: "))
      valido=False
      if edad >= 18 and edad <= 65:
        valido=True
      while not valido:
        print("La edad ingresada debe estar entre 18 y 65.")
        edad=int(input("Ingrese la edad del empleado: "))
        if edad >= 18 and edad <= 65:
          valido=True
      datos["edad"]=edad
      sector=input("Ingrese el sector del empleado (M: Mantenimiento, A: Administración, O: Operarios): ")
      sector=sector.upper()
      valido=False
      if sector=="M" or sector=="A" or sector=="O":
        valido=True
      while not valido:
        print("El sector ingresado no es válido.")
        sector=input("Ingrese el sector del empleado (M: Mantenimiento, A: Administración, O: Operarios): ")
        sector=sector.upper()
        if sector=="M" or sector=="A" or sector=="O":
          valido=True
      datos["sector"]=sector
      turno=input("Ingrese el turno del empleado (M o T): ")
      turno=turno.upper()
      valido=False
      if turno=="M" or turno=="T":
        valido=True
      while not valido:
        print("El turno ingresado no es válido.")
        turno=input("Ingrese el turno del empleado (M o T): ")
        turno=turno.upper()
        if turno=="M" or turno=="T":
          valido=True
      datos["turno"]=turno
      horas=int(input("Ingrese la cantidad de horas del empleado: "))
      valido=False
      if horas >= 10 and horas <= 40:
        valido=True
      while not valido:
        print("La cantidad de horas ingresada debe estar entre 10 y 40.")
        horas=int(input("Ingrese la cantidad de horas del empleado: "))
        if horas >= 10 and horas <= 40:
          valido=True
      datos["horas"]=horas
      lista_de_empleados.append(datos)
  if opcion=="2":
    contador_mañana=0
    contador_tarde=0
    for i in range(len(lista_de_empleados)):
      if lista_de_empleados[i]["turno"]=="M":
        contador_mañana+=1
      else:
        contador_tarde+=1
    print(f"La cantidad de empleados en el turno mañana es {contador_mañana} y la cantidad de empleados en el turno tarde es {contador_tarde}.")
  if opcion=="3":
    sector=input("Ingrese un sector (M: Mantenimiento, A: Administración, O: Operarios): ")
    sector=sector.upper()
    valido=False
    if sector=="M" or sector=="A" or sector=="O":
      valido=True
    while not valido:
      print("El sector ingresado no es válido.")
      sector=input("Ingrese nuevamente un sector (M: Mantenimiento, A: Administración, O: Operarios): ")
      sector=sector.upper()
      if sector=="M" or sector=="A" or sector=="O":
        valido=True
    suma=0
    contador=0
    for i in range(len(lista_de_empleados)):
      if lista_de_empleados[i]["sector"]==sector:
        suma+=lista_de_empleados[i]["horas"]
        contador+=1
    if contador==0:
      print(f"No hay empleados en el sector {sector}.")
    else:
      print(f"El promedio de horas en el sector {sector} es {suma/contador}.")
```

5. DICCIONARIOS CON LISTAS - Se desea almacenar los datos de 3 alumnos.

Definir un diccionario cuya clave sea el número de documento del alumno.

Como valor almacenar una lista con componentes de tipo tupla donde almacenamos nombre de materia y su nota.Tener en cuenta que un alumno puede cursar varias materias. Debe ir agregandole materias mientras el usuario lo desee.


Crear las siguientes opciones del menu:
1. Carga de los alumnos (de cada alumno solicitar su dni y los nombres de las materias y sus notas)
2. Listado de todos los alumnos con sus notas
3. Consulta de un alumno por su dni, mostrar las materias que cursa y sus notas.

```python
base_de_datos={}
opcion="0"
while opcion != "S":
  print()
  print("MENÚ:")
  print("1. Cargar un alumno")
  print("2. Listado de todos los alumnos con sus notas")
  print("3. Consulta de un alumno por su DNI")
  print("S. Salir")
  print()
  opcion=input("Ingrese una opción del menú: ")
  opcion=opcion.upper()
  print()
  if opcion=="1":
    datos=[]
    materias=[]
    notas=[]
    DNI=int(input("Ingrese el DNI del alumno: "))
    materia=input("Ingrese una materia: ")
    materias.append(materia)
    nota=float(input(f"Ingrese la nota de la materia {materia}: "))
    notas.append(nota)
    valido=False
    while not valido:
      otra=input("¿Desea agregar otra materia? s/n: ")
      otra=otra.upper()
      if otra == "N":
        valido=True
      else:
        materia=input("Ingrese una materia: ")
        materias.append(materia)
        nota=float(input(f"Ingrese la nota de la materia {materia}: "))
        notas.append(nota)
    for item in zip(materias,notas):
      datos.append(item)
    base_de_datos[DNI]=datos
  if opcion=="2":
    print(base_de_datos)
  if opcion=="3":
    DNI_a_mostrar=int(input("Ingrese el DNI del alumno a mostrar: "))
    valido=False
    for i in base_de_datos:
      if DNI_a_mostrar==i:
        valido=True
    if valido:
      print(base_de_datos[DNI_a_mostrar])
    else:
      print("El DNI del alumno a mostrar no está cargado en la base de datos.")
```
