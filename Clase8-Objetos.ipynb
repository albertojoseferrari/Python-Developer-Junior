*   Crear una clase de nombre fraccion que reciba como argumentos dos valores correspondientes al numerador y al denorminador de una fracción.
*   Crear un objeto de fracción y mostrar el numerador y el denominador.

**Ejercicio 2:**

* Crear una clase de nombre Gelatina.
La clase Gelatina recibirá como argumentos el tamaño en gramos, el color y el sabor.
*   Definir un método para mostrar las propiedades.
*   Crear tres objetos y mostrar las propiedades.

```python
class Gelatina(object):
  def __init__(self, tamanio, color, sabor):
    self.tamanio = tamanio
    self.color = color
    self.sabor = sabor

  def mostrar(self):
    print(f"El tamaño en gramos es {self.tamanio}, el color es {self.color} y el sabor es {self.sabor}.")

a = Gelatina(100,"rojo","dulce")
b = Gelatina(50,"verde","agrio")
c = Gelatina(200,"amarillo","ácido")

Gelatina.mostrar(a)
Gelatina.mostrar(b)
Gelatina.mostrar(c)
```

**Ejercicio 3:**

*   Modificar el ejercicio anterior para poder realizar la suma de dos fracciones de igual denominador. Mostrar el resultado.

```python
class Fraccion(object):
  def __init__(self, numerador, denominador):
    self.numerador = numerador
    self.denominador = denominador

  def __add__(self, otro):
    if self.denominador == otro.denominador:
      return Fraccion(self.numerador + otro.numerador, self.denominador)
    else:
      return "Las fracciones no tienen igual denominador."

  def __str__(self):
    return str(self.numerador) + "/" + str(self.denominador)

x = Fraccion(3,5)
y = Fraccion(6,5)
print(x+y)

x = Fraccion(3,5)
y = Fraccion(6,7)
print(x+y)
```

**Ejercicio**:

2. Crear una clase Coche que herede de Vehiculo
3. Crear una instancia de Coche que muestre algún atributo y comportamiento

```python
class Vehiculo(object):
  def __init__(self, largochasis, anchochasis, ruedas, enmarcha):
    self.largochasis = largochasis
    self.anchochasis = anchochasis
    self.ruedas = ruedas
    self.enmarcha = enmarcha

  def arrancar(self):
    self.enmarcha = True

  def estado(self):
    if self.enmarcha == True:
      return "El vehículo está en marcha."
    else:
      return "El vehículo está detenido."

class Coche(Vehiculo):
  def frenar(self):
    self.enmarcha = False

Ferrari = Coche(250,120,4,False)
print(Ferrari.ruedas)
Ferrari.arrancar()
Ferrari.estado()
```
