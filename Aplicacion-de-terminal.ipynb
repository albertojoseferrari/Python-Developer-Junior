 Se requiere:

 Una aplicación de terminal, con un menú que conste de las siguientes opciones:

1) Agregar Usuario

2) Actualizar Usuario

3) Borrar Usuario

4) Ver Base de Datos

S) Salir

Al usuario se le va a ofrecer ese menú y va a tener un prompt para poner la opción que busque.

En el menú de agregar usuarios se le va a pedir un nombre, el cual debe ser un valor de tipo string, un puesto de trabajo, y un email.
El puesto de trabajo debe ser uno de los siguientes:

Administrativo

Peón

Gerente

En caso de ingresarse otro valor, debe salir un mensaje que diga "Este es un valor incorrecto, recuerde que las opciones son Administrativo, Peón o Gerente"

El mail no hace falta que hagamos la comprobación ahora. Puede ser cualquier string, en el futuro vamos a aprender como se hace eso de que sea sí o sí un formato mail.

Además, el programa le va a asignar un identificador numérico, que debe ir creciendo cada vez que agreguemos un usuario.

Después, en el menú de actualizar usuario, debe pedir un id (el cual debe ser de nuestro usuario a modificar) y te va a pedir el nombre y que dejes vacío si no querés cambiarlo, te va a pedir un puesto de trabajo y lo mismo, si dejás vacío no lo cambia, y lo mismo con el mail. Luego debe decirte los datos actualizados y volver al menú.

En borrar sólo te va a pedir un usuario por id y lo va a borrar, devolviéndote el nombre del usuario borrado.

Y ver base de datos te va a mostrar la lista de usuarios ingresados.

```python
# inicializo la base de datos
base_de_datos = {}

# inicializo el id
id = 0

puestos_posibles = ["administrativo","peón","gerente"]
mensajes = {"ERROR_DATO_PUESTO": ("DATO INCORRECTO: Recuerde que los posibles valores son: " + " ".join(puestos_posibles)),
            "PEDIDO_DE_NOMBRE": "Ingrese el nombre del Usuario ",
            "PEDIDO_DE_PUESTO": "Ingrese el puesto de trabajo del Usuario ",
            "PEDIDO_DE_EMAIL": "Ingrese el email del Usuario ",
            "PEDIDO_DE_ID": "Ingrese el ID del Usuario "}

while True:
  print('''
        MENÚ:

        1) Agregar Usuario.
        2) Actualizar Usuario.
        3) Borrar Usuario.
        4) Ver Base de Datos.
        S) Salir.

        ''')
  opcion=input("Elija una opción: ").lower()

# Primera opción
  if opcion == "1":

    id = id+1

    nombre_ingresado = input(mensajes["PEDIDO_DE_NOMBRE"])

    while True:
      puesto_ingresado = input(mensajes["PEDIDO_DE_PUESTO"]).lower()

      # convierto peon en peón
      if puesto_ingresado == "peon":
        puesto_ingresado = "peón"

      if puesto_ingresado in puestos_posibles:
        # salgo del bulce si es correcto
        break
      else:
        print(mensajes["ERROR_DATO_PUESTO"])

    email_ingresado = input(mensajes["PEDIDO_DE_EMAIL"])

    # agrego el usuario creado
    base_de_datos[id] = [nombre_ingresado,puesto_ingresado,email_ingresado]

# Segunda opción
  elif opcion == "2":

    id_a_actualizar = int(input(mensajes["PEDIDO_DE_ID"] + "a actualizar "))

    nombre_a_actualizar = input(mensajes["PEDIDO_DE_NOMBRE"] + "a actualizar o deje vacío si no desea cambiarlo ")

    bandera_nombre = 0
    if nombre_a_actualizar != "":

      # registro que se actualizó el nombre
      bandera_nombre = 1

      # actualizo el nombre
      base_de_datos[id_a_actualizar][0] = nombre_a_actualizar

    while True:
      puesto_a_actualizar = input(mensajes["PEDIDO_DE_PUESTO"] + "a actualizar o deje vacío si no desea cambiarlo ").lower()

      # convierto peon en peón
      if puesto_a_actualizar == "peon":
        puesto_a_actualizar = "peón"

      if puesto_a_actualizar in puestos_posibles or puesto_a_actualizar == "":
        # salgo del bucle si el puesto es correcto (incluyendo la posibilidad de dejarlo vacío)
        break
      else:
        print(mensajes["ERROR_DATO_PUESTO"])

    bandera_puesto = 0
    if puesto_a_actualizar != "":

      # registro que se actualizó el puesto
      bandera_puesto = 1

      # actualizo el puesto
      base_de_datos[id_a_actualizar][1] = puesto_a_actualizar

    email_a_actualizar = input(mensajes["PEDIDO_DE_EMAIL"] + "a actualizar o deje vacío si no desea cambiarlo ")

    bandera_email = 0
    if email_a_actualizar != "":

      # registro que se actualizó el email
      bandera_email = 1

       # actualizo el email
      base_de_datos[id_a_actualizar][2] = email_a_actualizar

    if bandera_nombre == 1:
      print("Se ha actualizado el nombre del Usuario a", nombre_a_actualizar)

    if bandera_puesto == 1:
      print("Se ha actualizado el puesto de trabajo del Usuario a", puesto_a_actualizar)

    if bandera_email == 1:
      print("Se ha actualizado el email del Usuario a", email_a_actualizar)

# Tercera opción
  elif opcion == "3":

    id_a_eliminar = int(input(mensajes["PEDIDO_DE_ID"] + "a eliminar "))

    # guardo el nombre a eliminar
    nombre_a_eliminar = base_de_datos[id_a_eliminar][0]

    # elimino el usuario deseado
    base_de_datos.pop(id_a_eliminar)

    print("Se ha eliminado al Usuario", nombre_a_eliminar)

# Cuarta opción
  elif opcion == "4":

    print("La Base de Datos es", base_de_datos) # muestra la base de datos

# Salir
  elif opcion == "s":

    print("Usted eligió Salir.")
    # salgo del bucle del menú
    break

# Opción incorrecta
  else:
    print("La opción ingresada no es correcta.")
```
